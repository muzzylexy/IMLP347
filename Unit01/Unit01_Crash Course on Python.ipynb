### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
number = input()
number = int(number)
is_ok = False
if 1 <= number <= 1000:
    is_ok = True
if is_ok:
    if number % 2 == 0:
        print(number, "is even.")
    else:
        print(number, "is odd.")
else:
    print(f"{number} is out of range. The input number must be within range [1, 1000]")
````

### 02 - 填字遊戲（Mad libs game）
#### 使用者輸入任何字，它可以是名詞、形容詞、動詞、代名詞等。程式得到文字後，可自行排列，組成成一個模板故事。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一連串依據說明輸入對應文字 <br>
輸出(Output): 模板故事

* 進階: 可以讓程式重複執行

````python
def get_info_from_input(student_info):
    print("請輸入課程期數:", end='')
    id_course = int(input())
    print(id_course)
    student_info["id_course"] = id_course

    print("請輸入課程名稱:", end='')
    name_course = 0
    name_course = input()
    print(name_course)
    student_info["name_course"] = name_course
    
    print("請輸入姓名:", end='')
    name_student = 0
    name_student = input()
    print(name_student)
    student_info["name_student"] = name_student

    print("請輸入您的背景(科系或職業):", end='')
    identity_student = 0
    identity_student = input()
    print(identity_student)
    student_info["identity_student"] = identity_student
    
    print("請簡述對於本課程得期望(列出幾點修完課程" \
          "想得到的技能/對自己未來的規劃):", end='')
    purpose_student = 0
    purpose_student = input()
    print(purpose_student)
    student_info["purpose_student"] = purpose_student

def print_info(student_info):
    
    print("------------------------------------------")
    print("第", student_info["id_course"], "期 - ", student_info["name_course"], sep='')
    print("姓名: ", student_info["name_student"], sep='')
    print("學生背景: ", student_info["identity_student"], sep='')
    print("==========================================")
    print(student_info["name_student"], "修習完", student_info["name_course"], "後, ", sep='')
    print("結業成果獲得: ", student_info["purpose_student"], sep='')
    print("Well Done.", end="\n\n")

print("請問需要建造幾次模板故事:", end='')
num_requirements = 0
num_requirements = int(input())
print(num_requirements)

if num_requirements < 0:
    num_requirements = 0

student_info_list = list()

for count in range(0,num_requirements):
    student_info = dict()
    get_info_from_input(student_info)
    student_info_list.append(student_info)
    if count < num_requirements-1:
        print("請輸入下一筆資料")

for element in student_info_list:
    print_info(element)
````

### 03 -  猜數字（Guess the number）

````python
import random
inf = 0
sup = 10
ans = random.randint(inf+1, sup-1)
print(ans)
print("猜數字範圍 ", inf, " < ? < ", sup, ":", sep = "", end = " ")

count = 0
while True:
    guess = int(input())
    count += 1
    if guess > ans:
        print("再小一點, 您猜了", count, "次", sep="")
    elif guess < ans:
        print("再大一點, 您猜了", count, "次", sep="")
    else:
        print("賓果! 猜對了, 答案是", guess, sep="")
        break
````

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```
米花市帝丹小學一年级B班正舉辦期中考試
數學及格的有：柯南、灰原、步美、美環、光彦
英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
math_passed = {"柯南", "灰原", "步美", "美環", "光彦"}
english_passed = {"柯南", "灰原", "丸尾", "野口", "步美"}

all_student = math_passed | english_passed

print(f"所有學生名單: {all_student}")
print(f"數學及格但英文不及格的同學名單: {math_passed & (all_student - english_passed)}")
print(f"數學不及格但英文及格的同學名單: {(all_student - math_passed) & english_passed}")
print(f"兩者皆及格名單: {math_passed & english_passed}")
````

### 08 - 資料夾管理(shutil)
#### 說明 ####

```
在目前所在的目錄下建立一files資料夾

令使用者輸入一數字N，並在files資料夾中建立f1, f2… fN等N個資料夾後列出files的資料夾內容

將files資料夾裡的f1資料夾重新命名成folder1後再列出files的資料夾內容

移除files資料夾中的folder1後再列出files的資料夾內容

最後移除files資料夾
※須先退出files資料夾(os.chdir(../)) 
```


````python

import os
import shutil

def print_folder(in_folder):
    current_file_list = os.listdir(folder)
    print(in_folder, "內含檔案:")
    if current_file_list:
        for file in current_file_list:
            print(file)
    else:
        print("(無)")

folder = "files";

if not os.path.isdir(folder):
    os.mkdir(folder)

print("請輸入一數字:", end = "")
num_files = int(input())

for index in range(num_files, 0, -1):
    sub_folder = folder + "/f" + str(index)
    if not os.path.isdir(sub_folder):
        os.mkdir(sub_folder)

print_folder(folder)

for index in range(num_files, 0, -1):
    sub_folder = folder + "/f" + str(index)
    new_sub_folder = folder + "/folder" + str(index)
    try:
        os.rename(sub_folder, new_sub_folder)
    except:
        pass

print_folder(folder)

for index in range(num_files, 0, -1):
    sub_folder = folder + "/folder" + str(index)
    shutil.rmtree(sub_folder)

print_folder(folder)
shutil.rmtree(folder)
````

### 09 - 回文判斷（Is a palindrome）
#### 請使用者輸入單字，判斷它是否為回文，也就是該單字前後對稱，例如 madam，從前讀到後或是從後讀到前的順序都是 madam。

> 題目：<br>
輸入(Input): 雨 滋 春 樹 碧 連 天  天 連 碧 樹 春 滋 雨 <br>
輸出(Output):The text you have entered is a palindrome!<br>
輸入(Input): 資工訓練班 <br>
輸出(Output):The text you have entered is not a palindrome.


````python
import string

sentence = input("請輸入句子: ")
sentence = sentence.lower()
sentence = sentence.translate({ord(letter):"" for letter in string.punctuation})

print(sentence)

num_checking = len(sentence) // 2
reversed_sentence = sentence[::-1]

for index in range(0, num_checking):
    if sentence[index] != reversed_sentence[index]:
        print("The text you have entered is not a palindrome.")
        break
else:
    print("The text you have entered is a palindrome!")

````

### 10 - 總成績計算
#### 說明 ####

請讀取 `data/english_list.csv`與`data/math_list.csv`檔案(`utf-8`)後，將每位同學的英文與數學成績加總起來，並將檔案寫入至 `./Score.csv`中，編碼成`utf-8`<br>
最後列印出每位學生姓名與加總後的分數<br>
<a href="data\english_list.csv">英文成績下載</a><br>
<a href="data\math_list.csv">數學成績下載</a><br>

#### Input Format ####

<br>
csv資料(英文成績)<br>
csv資料(數學成績)<br>
<br>

#### Output Format ####

<br>

<br>
<br>
(圖片參考用 以文字敘述為準)<br>




#### Sample Input 1####




#### Sample Output 1####


```
廖冠霖 142
王力中 124

.
陳姿茜 77
涂珮瑜 134
夏明哲 111

```


#### Hint ####



```
None
```


````python
import copy
import string
import os

def read_csv(filename):
    if not os.path.exists(filename):
        return dict()
    file = open(filename, "r", encoding = "UTF-8")
    file.readline()
    data_pairs = dict()
    while True:
        line = file.readline()
        if line:
            # skip lines with no data
            line = line.translate({ord(letter):"" for letter in string.whitespace})
            if not line:
               continue
            key, separator, raw_value = line.partition(",")
            try:
                value = int(raw_value)
            except:
                value = 0
            data_pairs[key] = int(value)
        else:
            break
    file.close()
    return data_pairs


def write_csv(filename, data_pairs):
    file = open(filename, "wt", encoding = "UTF-8")
    for name, score in data_pairs.items():
        print(f"{name},{score}", file = file)

english_score_data = read_csv("./data/english_list.csv")
math_score_data = read_csv("./data/math_list.csv")
summation_score_data = copy.deepcopy(english_score_data)

for name, math_score in math_score_data.items():
    if name in summation_score_data:
        summation_score_data[name] += math_score
    else:
        summation_score_data[name] = math_score

write_csv("./Score.csv", summation_score_data)

file = open("./Score.csv", "rt", encoding = "UTF-8")
print(file.read())
````
